```python
import numpy as np
import matplotlib.pyplot as plt

def larsen_vs_mamdani_example():
    print("імплікація: Ларсен проти Мамдані")

    #вісь Х
    x = np.linspace(0, 100, 100)

    #Створюємо функція належності "Швидко"
    fast_speed = np.maximum(0, (x - 50) / 50)

    #Припустимо, правило спрацювало з істинністю 0.6
    alpha = 0.6
    print(f"Ступінь істинності правила (alpha) = {alpha}\n")

    #ІМПЛІКАЦІЯ МАМДАНІ
    #Зрізаємо все, що вище 0.6
    mamdani_result = np.minimum(alpha, fast_speed)

    #ІМПЛІКАЦІЯ ЛАРСЕНА
    #Множимо кожну точку графіка на 0.6
    larsen_result = alpha * fast_speed

    print("Ларсен зберігає форму трикутника,")
    print("тоді як Мамдані перетворює його на трапецію (зрізає пік).")

    #результати
    plt.figure(figsize=(10, 5))

    # Оригінальний трикутник
    plt.plot(x, fast_speed, 'k--', label='Оригінальна множина "Швидка"', alpha=0.5)

    # Результат Мамдані (синя лінія)
    plt.plot(x, mamdani_result, 'b-', linewidth=2, label='Мамдані (Мінімум - Зрізання)')
    plt.fill_between(x, 0, mamdani_result, color='blue', alpha=0.1)

    # Результат Ларсена (червона лінія)
    plt.plot(x, larsen_result, 'r-', linewidth=2, label='Ларсен (Добуток - Масштабування)')
    plt.fill_between(x, 0, larsen_result, color='red', alpha=0.1)

    plt.title('Порівняння алгоритмів при рівні істинності $\\alpha = 0.6$')
    plt.xlabel('Швидкість обертання (%)')
    plt.ylabel('Ступінь належності $\\mu(y)$')
    plt.ylim(0, 1.1)
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    larsen_vs_mamdani_example()
```
