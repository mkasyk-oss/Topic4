```python
import numpy as np

def zadeh_composition_example():
    print("Алгоритм узагальнення Заде (Максимінна композиція)")

    # нечіткі множини для правила
    A = np.array([1.0, 0.6, 0.2])

    B = np.array([0.2, 0.7, 1.0])

    #нечітке відношення R (ЯКЩО A ТО B)
    #(імплікація Заде)
    R = np.minimum.outer(A, B)
    print("Матриця відношення R (Правило):")
    print(R)
    print("-" * 40)

    #нове спостереження A'
    #(трохи гірше, ніж ідеал A)
    A_prime = np.array([0.8, 0.9, 0.4])
    print(f"Нове спостереження A': {A_prime}")

    #алгоритм Заде (B' = A' ◦ R)
    #B'(y) = max( min( A'(x), R(x,y) ) )
    B_prime = np.zeros(len(B))

    for y in range(len(B)):
        R_column = R[:, y]
        min_values = np.minimum(A_prime, R_column)
        B_prime[y] = np.max(min_values)

    print(f"Результат B' (Оцінка надійності): {B_prime}")
    print("-" * 40)

if __name__ == "__main__":
    zadeh_composition_example()
```
